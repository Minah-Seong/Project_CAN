```python
import requests
import json
import pandas as pd
import numpy as np
from pandas.io.json import json_normalize
import os
import webbrowser
import folium
from folium import plugins
print(folium.__version__)
```

```python
geo_path = 'data/전국.json'
try:
    geo_data1 = json.load(open(geo_path, encoding='utf-8'))
except:
    geo_data1 = json.load(open(geo_path, encoding='utf-8-sig'))
```

```python
df = pd.read_excel("data/전국 전세가율 지도코드 데이터.xlsx")
df["코드"]=df["코드"].astype(str)
df
```

```python
state_geo = 'data/전국.json'
state_geo2 = json.load(open(state_geo, encoding ="utf-8"))
```

```python
m = folium.Map(location = [35.8, 127.6], tiles="OpenStreetMap", zoom_start=8)

choropleth = folium.Choropleth(
    geo_data=state_geo2,
    name = "지도",
    data = df,
    columns=('SIG_KOR_NM', "전세가율"),
    fill_color='YlOrRd',
    key_on="feature.properties.SIG_KOR_NM",
    fill_opactiy = 1,
    line_opacity = 0.5,
    nan_fill_color='#FAEB78',
    nan_fill_opacity=0.1,
).add_to(m)


```

```python
m

m.save('지도 시각화.html')
```
