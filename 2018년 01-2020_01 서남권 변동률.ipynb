```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import sys
from matplotlib import font_manager, rc, cm
mpl.rcParams['axes.unicode_minus'] = False

font_path = './font/GmarketSansTTFMedium.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
data = pd.read_excel('./data/서남권_월간_매매가격지수_연립다세대.xlsx')
data.set_index('지 역' ,inplace=True)
data
```

```python
# colors = sns.color_palette('BuGn',len(ndf.columns))

fig = plt.figure(figsize=(12,6))
fig.set_facecolor('white')
ax = fig.add_subplot()


line1 = ax.plot(data.columns, data.loc['서남권'], label='서남권', linewidth=5, marker='o', markersize=10)
plt.hlines(0,'2018년 09월','2019년 11월',color="grey")


ax.set_title('2018.09 ~ 2019.11 서남권 매매가변동지수 시계열 그래프',size=20, pad=10)

ax.set_xticklabels(labels=data.columns, fontsize=13)

plt.xticks([0,2,4,6,8,10, 12, 14], ['2018년 09월', '2018년 11월', '2019년 01월', '2019년 03월', 
                                    '2019년 05월', '2019년 07월','2019년 09월','2019년 11월'])
plt.yticks(fontsize = 13)

ax.set_ylabel('전월대비 매매가 변동률', fontsize = 15)
ax.set_xlabel('월', fontsize = 15)
ax.legend(loc='best')

plt.savefig('output/2018.09 ~ 2019.11 서남권_매매가변동지수_시계열_그래프.png')
```
