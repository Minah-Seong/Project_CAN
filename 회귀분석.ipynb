```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import sys
from matplotlib import font_manager, rc, cm                
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split


font_path = './font/GmarketSansTTFMedium.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
df = pd.read_csv('./data/통합파일_군집.csv', index_col=0)
df
df.corr()
```

```python
clu1 = df[(df['cluster'] == 0)]
clu2 = df[(df['cluster'] == 1)]
clu3 = df[(df['cluster'] == 2)]
```

```python
# 통합 회귀분석

X = df[['건축허가면적(m2)','지하철역 수(개)','1인가구(세대)','아파트 매매거래 수','아파트 입주예정 물량(세대)','직장인연봉(만원)','직장인수(명)']]
Y = df['전세가율(%)']

# X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15)
# print(X_train.shape)
# print(X_test.shape)

lr = LinearRegression()

lr.fit(X,Y)

r_square = lr.score(X, Y)
print(r_square)

print(lr.coef_)

print(lr.intercept_)
```
