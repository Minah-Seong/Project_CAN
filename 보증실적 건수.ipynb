```python
import pandas as pd
import seaborn as sns
import re
import matplotlib
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import sys
from matplotlib import font_manager, rc, cm                ## 폰트를 관리하는 함수와 설정 함수를 사용한다. 

font_path = './font/GmarketSansTTFMedium.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
df = pd.read_csv("./data/갭투자비율및깡통전세비율.csv", encoding = "CP949", index_col= "지역")
df.columns = ["2021년 갭투자 건수", "2021년 깡통전세 건수", "2020년 갭투자 건수", "2020년 깡통전세 건수"]
df = df.drop("전국", axis = 0)
df
```

```python
df1 = df.sort_values(by='2021년 갭투자 건수', ascending=False)
df1 = df1.head(7)
```

```python
fig, ax = plt.subplots(figsize=(15,6))
bar_width = 0.25

index = np.arange(7)
location = df1.index

b2 = plt.bar(index, df1['2021년 갭투자 건수'], bar_width, alpha=0.4, color='blue', label='2021년 갭투자 건수')

b3 = plt.bar(index + bar_width, df1['2020년 갭투자 건수'], bar_width, alpha=0.4, color='green', label='2020년 갭투자 건수')

# x축 위치를 정 가운데로 조정하고 x축의 텍스트를 year 정보와 매칭
plt.xticks(np.arange(bar_width, 7 + bar_width, 1), location)

# x축, y축 이름 및 범례 설정
plt.title('갭투자 건수')
plt.xlabel('지역', size = 13)
plt.ylabel('갭투자 건수', size = 13)
plt.legend()
plt.show()
```

```python
df2 = df.sort_values(by='2021년 깡통전세 건수', ascending=False)
df2 = df2.head(7)
```

```python
fig, ax = plt.subplots(figsize=(15,6))
bar_width = 0.25

index = np.arange(7)
location = df2.index

b2 = plt.bar(index, df2['2020년 깡통전세 건수'], bar_width, alpha=0.4, color='blue', label='2020년 깡통전세 건수')

b3 = plt.bar(index + bar_width, df2['2021년 깡통전세 건수'], bar_width, alpha=0.4, color='green', label='2021년 깡통전세 건수')

# x축 위치를 정 가운데로 조정하고 x축의 텍스트를 year 정보와 매칭
plt.xticks(np.arange(bar_width, 7 + bar_width, 1), location)

# x축, y축 이름 및 범례 설정
plt.xlabel('지역', size = 13)
plt.ylabel('깡통전세 건수', size = 13)
plt.legend()
plt.show()
```
