```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import sys
from matplotlib import font_manager, rc, cm                ## 폰트를 관리하는 함수와 설정 함수를 사용한다. 

font_path = './font/GmarketSansTTFMedium.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
df = pd.read_excel("data/강서구일대 사기피해현황.xlsx",index_col=1)
df.drop('지 역', axis=1, inplace=True)
df
```

```python
ndf = df[['2017건수','2018건수','2019건수','2020건수','2021건수']]
price = df[['2017금액','2018금액','2019금액','2020금액','2021금액']]
ndf.fillna(0, inplace=True)
price.fillna(0, inplace=True)

```

```python
ndf = ndf.T
price = price.T

ndf
```

```python
# colors = sns.color_palette('BuGn',len(ndf.columns))

fig = plt.figure(figsize=(10,8))
fig.set_facecolor('white')
ax = fig.add_subplot()

line1 = ax.plot(ndf.columns, ndf.loc['부천시'], label='경기도 부천시', linewidth=4, alpha=0.7, color='Blue', marker='o', markersize=10)
line1 = line1[0]

line2 = ax.plot(ndf.columns, ndf.loc['강서구'], label='서울시 강서구', linewidth=4, alpha=0.7, color='Green', marker='o', markersize=10)
line2 = line2[0]

line3 = ax.plot(ndf.columns, ndf.loc['양천구'], label='서울시 양천구', linewidth=4, alpha=0.7, color='Red', marker='o', markersize=10)
line3 = line3[0]

xylist1 = list(line1.get_xydata())
for coord in xylist1[2:3]:
    ax.text(coord[0]+0.2,coord[1]-15,f'{round(coord[1])}건',fontsize=13, ha='center')
for coord in xylist1[3:4]:
    ax.text(coord[0],coord[1]-35,f'{round(coord[1])}건',fontsize=13, ha='center')
for coord in xylist1[4:5]:
    ax.text(coord[0],coord[1]+20,f'{round(coord[1])}건',fontsize=13, ha='center')

xylist2 = list(line2.get_xydata())
for coord in xylist2[2:5]:
    ax.text(coord[0]-0.08,coord[1]+23,f'{round(coord[1])}건',fontsize=13, ha='center')
    
xylist3 = list(line3.get_xydata())
for coord in xylist3[2:5]:
    ax.text(coord[0]-0.05,coord[1]+20,f'{round(coord[1])}건',fontsize=13, ha='center')

ax.set_title('강서구 일대 전세보증금반환보증 사고현황',size=20, pad=10)


ax.set_xticklabels([2017,2018,2019,2020,2021], fontsize=11)
ax.set_ylim(0,630)
ax.set_xlabel('연도', fontsize = 15)
ax.set_ylabel('사고건수(건)', fontsize = 15)
               
ax.legend(loc='upper left')

plt.savefig('output/강서구_일대_전세보증금반환보증_사고현황.png')
```
