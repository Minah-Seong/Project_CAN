```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sys
from matplotlib import font_manager, rc, cm

font_path = './font/GmarketSansTTFMedium.ttf'
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
data = pd.read_excel('./data/전세보증금반환보증 사고현황.xlsx')
data = data.set_index('연도')

```

```python
data.plot(linewidth=4, marker='o', markersize=10, figsize=(8,6))
plt.title('지역별 전세보증금반환보증 사고건수 현황', fontsize=15)
plt.ylabel('사고건수(건)',size=13)
plt.xlabel('연도',size=13)

plt.savefig('./output/지역별_전세보종금반환보증_사고건수 현황.png')
```
